```python
import pandas as pd

train_data =  pd.read_csv('data/cleaned_train_iffat.csv')
test_data =  pd.read_csv('data/test.csv')
train_data
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Split the data into training, validation, and test sets
X_train, X_val, y_train, y_val = train_test_split(train_data.drop(columns=["y","UID","col_4"]), train_data["y"], test_size=0.2, random_state=42)

# Define a list of regression models to compare
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Gradient Boosting Regressor': GradientBoostingRegressor()
}

# Initialize variables to store model performance
best_model = None
best_rmse = float('inf')

# Model selection loop
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    
    print(f"{model_name} RMSE: {rmse}")
    
    if rmse < best_rmse:
        best_rmse = rmse
        best_model = model_name

print(f"Best Model: {best_model} (RMSE: {best_rmse})")

# Evaluate the best model on the test data
# best_model = models[best_model]
# y_test_pred = best_model.predict(X_test)
# test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
# print(f"Test RMSE of Best Model: {test_rmse}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor,
    AdaBoostRegressor,
)
import xgboost as xgb
import lightgbm as lgb
import catboost as catb
import random

# Split the data into training, validation, and test sets
X_train, X_val, y_train, y_val = train_test_split(train_data.drop(columns=["y","UID","col_4"]), train_data["y"], test_size=0.1, random_state=int(random.random()*10+40))

# Define a list of regression models to compare
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Gradient Boosting Regressor': GradientBoostingRegressor(),
    'AdaBoost Regressor': AdaBoostRegressor(),
    'XGBoost Regressor': xgb.XGBRegressor(),
    'CatBoost Regressor': catb.CatBoostRegressor(silent=True),
    'LightGBM Regressor': lgb.LGBMRegressor(silent=True),
}

# Initialize variables to store model performance
best_model = None
best_rmse = float('inf')

# Model selection loop
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    
    print(f"{model_name} RMSE: {rmse}")
    
    if rmse < best_rmse:
        best_rmse = rmse
        best_model = model_name

print(f"Best Model: {best_model} (RMSE: {best_rmse})")


# Evaluate the best model on the test data
# best_model = models[best_model]
# y_test_pred = best_model.predict(X_test)
# test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
# print(f"Test RMSE of Best Model: {test_rmse}")
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

# Define hyperparameter grids for each model
param_grids = {
    'Linear Regression': {},
    'Ridge Regression': {'alpha': [0.1, 1.0, 10.0]},
    'Lasso Regression': {'alpha': [0.1, 1.0, 10.0]},
    'Decision Tree Regressor': {'max_depth': [None, 10, 20, 30]},
    'Random Forest Regressor': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20, 30]},
    'Gradient Boosting Regressor': {'n_estimators': [50, 100, 200], 'max_depth': [3, 4, 5]},
    'AdaBoost Regressor': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1.0]},
    'XGBoost Regressor': {'n_estimators': [50, 100, 200], 'max_depth': [3, 4, 5]},
    'LightGBM Regressor': {'n_estimators': [50, 100, 200], 'max_depth': [3, 4, 5]},
    'CatBoost Regressor': {'iterations': [50, 100, 200], 'depth': [4, 6, 8]},
}

# Initialize variables to store best models and their RMSE scores
best_models = {}
best_rmse_scores = {}

# Hyperparameter tuning loop
for model_name, model in models.items():
    param_grid = param_grids[model_name]
    
    if param_grid:
        search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=5)
        search.fit(X_train, y_train)
        
        best_model = search.best_estimator_
        best_rmse = np.sqrt(-search.best_score_)
    else:
        model.fit(X_train, y_train)
        best_model = model
        y_pred = model.predict(X_val)
        best_rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    
    best_models[model_name] = best_model
    best_rmse_scores[model_name] = best_rmse
    print(f"Best {model_name} RMSE: {best_rmse}")

# Find the model with the lowest RMSE
best_model_name = min(best_rmse_scores, key=best_rmse_scores.get)
print(f"Best Model: {best_model_name} (RMSE: {best_rmse_scores[best_model_name]})")

# Evaluate the best model on the test data
# best_model = best_models[best_model_name]
# y_test_pred = best_model.predict(X_test)
# test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
# print(f"Test RMSE of Best Model: {test_rmse}")
```
